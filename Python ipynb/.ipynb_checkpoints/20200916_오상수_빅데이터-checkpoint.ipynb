```python
import numpy as np
import pandas as pd
```

```python
daegu=pd.read_excel('data/6270000_대구광역시_07_24_04_P_일반음식점.xlsx', encoding='utf-8')
daegu.head()
```

```python
daegu.columns
```

```python
dg=daegu.loc[daegu.상세영업상태명=='영업']
dg
```

```python
dg.notnull().sum()
```

```python
dg1=dg.dropna(thresh=30510, axis=1)
dg1.head()
```

```python
dg1.columns
```

```python
dg1.업태구분명.unique()
```

```python
dg1=dg1.dropna(subset=['업태구분명'])
dg1.head()
```

```python
ck_name=pd.Series(dg1.업태구분명.unique())
ck_name
```

```python
ck=ck_name[ck_name.str.contains('통닭')]
ck
```

```python
dg1=dg1.loc[dg1.업태구분명.isin(ck)]
dg1.head()
```

```python
dg1.columns
```

```python
dg1.소재지전체주소
```

```python
dg1.소재지전체주소.isnull().sum()
```

```python
dg1=dg1.dropna(subset=['소재지전체주소'])
dg1.소재지전체주소.isnull().sum()
```

```python
dg2=dg1.copy()
dg2[['광역시', '구', '동네', '나머지주소']] = dg1['소재지전체주소'].str.split(' ',n=3, expand=True)
dg2.head()
```

```python
dg2.동네.unique()
```

```python
dgcount1=dg2.groupby('동네').count()
dgcount1.head()
```

```python
dgcount=dg2.groupby('동네').나머지주소.count()
dgcount
```

```python
dg2.
```

```python
countmax=dgcount.values.max()
countmax
```

```python
dgcount[dgcount.values==countmax]
```

```python
dg2.구.unique()
```

```python
dg2.구.count()
```

```python
dgcount2=dg2.groupby('구').나머지주소.count()
dgcount2
```

```python
label=dgcount2.index
print(label)
index = np.arange(len(label))
print(index)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'

plt.figure(figsize=(10,10))
plt.bar(index,dgcount2)
plt.title('구별 치킨집의 수')
plt.xticks(index,label)
plt.show()
```
