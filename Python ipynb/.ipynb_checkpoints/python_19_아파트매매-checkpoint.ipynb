```python
import numpy as np
import pandas as pd
```

```python
file_path='data/아파트(매매)__실거래가_달서구.csv'
df = pd.read_csv(file_path, encoding='euc-kr', skiprows=15, thousands = ',')
```

```python
file_path='data/아파트(매매)__실거래가_수성구.csv'
df2 = pd.read_csv(file_path, encoding='euc-kr', skiprows=15, thousands = ',')
```

```python
dal=df.copy()
soo=df2.copy()
```

```python
mae=pd.concat([df,df2])
mae.head()
```

```python
df=mae.copy()
```

```python
df=df[['시군구', '번지', '단지명', '전용면적(㎡)', '거래금액(만원)']]
df.head()
```

```python
df=df.astype({'전용면적(㎡)':int})
df.dtypes
```

```python
df=df.sort_values(by='전용면적(㎡)', ascending=False)
df.head()
```

```python
df[['시도', '구', '나머지']]=df['시군구'].str.split(' ', n=2, expand=True)
df.head()
```

```python
drop_list2 = ['시군구', '시도', '나머지']
df = df.drop(drop_list2, axis=1)
df.head()
```

```python
df['count']=1
df
```

```python
m2=df['count'].groupby([df['전용면적(㎡)']]).sum().sort_values(ascending=False)
m2.head()
```

```python
m2[m2.values==m2.max()]
```

```python
dfm2=df[df['전용면적(㎡)']==84]
dfm2.head()
```

```python
# df=df.astype({'거래금액(만원)':int})
dfm2.dtypes
```

```python
soomean=dfm2['거래금액(만원)'].groupby([dfm2['구'],dfm2['단지명']]).mean().round()
soomean
```

```python
print(soomean['수성구'].values.max())
print(soomean['수성구'].values.min())
```

```python
print(soomean['수성구'][soomean['수성구'].values==soomean['수성구'].values.max()])
print(soomean['수성구'][soomean['수성구'].values==soomean['수성구'].values.min()])
```

```python
print(soomean['달서구'].values.max())
print(soomean['달서구'].values.min())
```

```python
print(soomean['달서구'][soomean['달서구'].values==soomean['달서구'].values.max()])
print(soomean['달서구'][soomean['달서구'].values==soomean['달서구'].values.min()])
```

```python
top7 = soomean['수성구'].sort_values(ascending=False)[:7]
top7
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = 'C:/Windows/Fonts/NGULIM.TTF'
font_family = fm.FontProperties(fname=font_path).get_name()
plt.rcParams["font.family"] = font_family
# 위코드는 한글지원 폰트 설정

top7_values = list(np.ravel(top7.values))
x = np.arange(len(top7_values))  # the label locations
width = 0.6  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 7)) #fig, ax = plt.subplots(TOTAL_X,TOTAL_Y,figsize=(8, 5))
plt.grid(b=True, which='both', axis='both')
rects1 = ax.bar(top7.index, top7_values, width, align='center', alpha=0.7, label='평균거래금액(만)')
rects1[top7_values.index(max(top7_values))].set_color('g')

ax.set_title('수성구 아파트 평균거래금액 Top7')
ax.set_xticks(x)
ax.set_xticklabels(top7.index)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
fig.tight_layout()

plt.show()
```
