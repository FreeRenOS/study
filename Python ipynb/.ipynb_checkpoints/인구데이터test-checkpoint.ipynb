```python
import pandas as pd
import numpy as np
```

## 1. 10살 단위로 인구를 정리한다

```python
resident1 = pd.read_csv('data/2018.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv')
resident1.head()
```

```python
df = resident1.copy()
```

```python
drop_list = []
check_list = ['총계', '구분', ' 남', ' 여'] # 남구 삭제 방지를 위해 '남' 앞에 공백 추가 
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용
```

```python
df1= df.drop(drop_list, axis=1)
df1
```

```python
df1['인구0'] = (pd.DataFrame(range(0,101))/10).astype(int)*10
df1 = df1.groupby('인구0')
df1 = df1.sum()
df1
```

## 2. 70대 이상은 하나로 합친다.

```python
df1['인구'] = df1.index
df1['인구'] = np.where(df1['인구'] >= 70, 70, df1.인구)
df1 = df1.groupby('인구')
df1 = df1.sum()
df1
```

```python
resident2 = pd.read_csv('data/2018.12.31.기준 대구광역시 주민등록인구 현황_UTF8.csv')
resident2.head()
```

```python
df = resident2.copy()

drop_list = []
check_list = ['총계', '구분', ' 남', ' 여'] # 남구 삭제 방지를 위해 '남' 앞에 공백 추가 
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용

df2= df.drop(drop_list, axis=1)
df2
```

```python
df2['인구'] = (pd.DataFrame(range(0,101))/10).astype(int)*10
df2['인구'] = np.where(df2['인구'] >= 70, 70, df2.인구)
df2 = df2.groupby('인구')
df2 = df2.sum()
df2
```

```python
resident3 = pd.read_csv('data/2019.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv', thousands=',')

df = resident3.copy()

drop_list = []
check_list = ['총계', '행정구역', ' 남', ' 여'] # 남구 삭제 방지를 위해 '남' 앞에 공백 추가 
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용

df3= df.drop(drop_list, axis=1)

df3['인구'] = (pd.DataFrame(range(0,101))/10).astype(int)*10
df3['인구'] = np.where(df3['인구'] >= 70, 70, df3.인구)
df3 = df3.groupby('인구')
df3 = df3.sum()
df3
```

## 3. 3개의 기간동안 인구가 증가하는 곳을 찾는다

```python
# 전치
df1 = df1.T
df1['총합']= df1.sum(axis='columns').astype(int)
df1
```

```python
df2 = df2.T
df2['총합']= df2.sum(axis='columns').astype(int)
df2
```

```python
df3 = df3.T
df3['총합']= df3.sum(axis='columns').astype(int)
df3
```

```python
pop1 = (df2['총합'] - df1['총합']) > 0
pop2 = (df3['총합'] - df2['총합']) > 0
df3['인구증가'] = pop1 & pop2
df3
```

```python
df3[df3.인구증가 == True]
```
