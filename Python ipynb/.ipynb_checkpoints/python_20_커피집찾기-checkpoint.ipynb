```python
import math
def dis(lat,lon) :
    Seoul = lat  
    Toronto = lon
    lat1 = Seoul[0]
    lon1 = Seoul[1]
    lat2 = Toronto[0]
    lon2 = Toronto[1]
    #Latitude, Longitude
    
    R = 6371000 # metres
    #위도(latitude), 경도(longitude)

    rad_lat1 = lat1 * math.pi*(1/180) # latitude, longitude in radians (angle to radians)
    rad_lat2 = lat2 * math.pi*(1/180)
    dis_lat = (lat2-lat1) * math.pi*(1/180)
    dis_lon = (lon2-lon1) * math.pi*(1/180)

    a = math.sin(dis_lat/2) * math.sin(dis_lat/2) + math.cos(rad_lat1) * math.cos(rad_lat2) * math.sin(dis_lon/2) * math.sin(dis_lon/2) #a = sin²(Δφ/2) + cos φ1 ⋅ cos φ2 ⋅ sin²(Δλ/2)

    b= 1-a
    c = 2 * (math.atan2( math.sqrt(a), math.sqrt(b)))

    d = R * c #in metres
    
    return d
```

```python
dislist=[]
for i in stlist :
    for j in edlist :
        dislist.append(dis(i,j))
np.min(dislist)
```

```python
for i in range(len(dislist2)) :
    stdisindex=int(dislist2[i][0]/len(edlist))
    st_T=pd.DataFrame(st.iloc[stdisindex]).T

    edisindex=dislist2[i][0]%len(edlist)
    ed_T=pd.DataFrame(ed.iloc[edisindex]).T

    result_g=pd.concat([st_T, ed_T], ignore_index=True)
    result_g['그룹']=i
    result_g['거리(m)']=round(dislist2[i][1],1)
    
    result=pd.concat([result,result_g],ignore_index=True)
result
```

```python
# 지도 시각화를 위한 라이브러리
import folium
```

```python
lat_m = result["위도"].mean()
lng_m= result["경도"].mean()
lat_m, lng_m
```

```python
m = folium.Map([lat_m, lng_m], zoom_start=12) #, tiles="Stamen Toner")

for i in result.index:
    sub_lat = result.loc[i, "위도"]
    sub_lng = result.loc[i, "경도"]
    
    title = f"{result.loc[i, '상호명']} - {result.loc[i, '지번주소']}"
    
    color = "green"
    if result.loc[i, "브랜드"] == "이디야":
        color = "blue"
    
    folium.CircleMarker([sub_lat, sub_lng],
                        radius=3,
                        color=color,
                          tooltip=title).add_to(m)
```
